# Comparison of value and policy iteration algorithm (Reinforcement learning)

## Imports

```python
from value_iteration import valueIteration
from policy_iteration import policyIteration
import matplotlib.pyplot as plt
import numpy as np
```

## Computation

```python
thetas = np.array([1e-1, 1e-5, 1e-10, 1e-15, 1e-20, 1e-25, 1e-30, 1e-35, 1e-40])
gamma = 1
living_reward, goal_reward, death_reward = 0.05, 1, -1
r = [living_reward, goal_reward, death_reward]
tv, epv, tp, epp = [], [], [], []
```

```python
for theta in thetas:
    ep1, t1 = valueIteration(gamma, theta, r, silent = True)
    tv.append(t1)
    epv.append(ep1)
    ep2, t2  = policyIteration(gamma, theta, r, silent = True)
    tp.append(t2)
    epp.append(ep2)
```

## Plots

```python
plt.scatter(np.log(thetas), epv)
plt.scatter(np.log(thetas), epp)
plt.title('Number of episodes vs theta')
plt.xlabel = "Log of theta"
plt.ylabel = "# Episodes"
plt.legend(["Value iteration", "Policy iteration"])
```

```python
plt.scatter(np.log(thetas), tv)
plt.scatter(np.log(thetas), tp)
plt.xlabel = "Log of theta"
plt.ylabel = "Time"
plt.title("Runtime vs theta")
plt.legend(["Value iteration", "Policy iteration"])
```

# Observations

- **Value iteration**: Takes more time but less number of episodes to converge
- **Policy iteration**: Takes less time but more number of episodes to converge
