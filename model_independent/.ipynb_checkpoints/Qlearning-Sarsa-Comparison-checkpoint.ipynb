# Imports

```python
from sarsa import sarsa
from qlearning import qlearning
import matplotlib.pyplot as plt
```

# Qlearning

```python
alpha = 0.4
gamma = 0.999
epsilon = 0.9
episodes = 2000
max_steps = 2500
n_tests = 20
q_timestep_reward = qlearning(alpha, gamma, epsilon, episodes, max_steps, n_tests)
```

```python
plt.plot(q_timestep_reward)
```

# Sarsa

```python
alpha = 0.4
gamma = 0.999
epsilon = 0.9
episodes = 2000
max_steps = 2500
n_tests = 20
timestep_reward = sarsa(alpha, gamma, epsilon, episodes, max_steps, n_tests)
print(timestep_reward)
```
